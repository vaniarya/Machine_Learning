```python
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('weatherAUS.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
year = pd.to_datetime(df.Date).dt.year

train_df = df[year < 2015]
val_df = df[year == 2015]
test_df = df[year > 2015]
```

```python
inputs = list(train_df)[1:-1]
targets = 'RainTomorrow'
train_input = train_df[inputs]
train_target = train_df[targets]
```

```python
test_input = test_df[inputs]
test_target = test_df[targets]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
train_target = le.fit_transform(train_target)
```

```python
pipe = Pipeline([
    ('encode', OneHotEncoder(sparse_output=False, handle_unknown='ignore')),
    ('impute', SimpleImputer()),
    ('scale' , MinMaxScaler()), 
    ('model' , DecisionTreeClassifier(random_state = 42))
])
```

```python
mod = GridSearchCV(estimator = pipe,
                  param_grid = {},
                  cv= 3,
                  scoring = 'accuracy')
```

```python
mod.fit(train_input, train_target)
```
