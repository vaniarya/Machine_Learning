```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
num_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
```

```python
from sklearn.decomposition import PCA
```

```python
?PCA
```

```python
pca = PCA(n_components = 2)
```

```python
pca.fit(iris[num_cols])
```

```python
transformed = pca.transform(iris[num_cols])
```

```python
transformed
```

```python
sns.scatterplot(x=transformed[:,0], y=transformed[:,1], hue=iris['species']);
```

```python
from sklearn.manifold import TSNE
```

```python
?TSNE
```

```python
tsne = TSNE(n_components=2)
```

```python
transformed2 = tsne.fit_transform(iris[num_cols])
```

```python
transformed2
```

```python
sns.scatterplot(x=transformed2[:,0], y=transformed2[:,1], hue=iris['species']);
```
